## Filering based on the column

```python
df[df['Rank'] <10]
```

## Filtering based on specific values in the column 

```python
df[df['Rank']== 5] # takes a single value only
```

```python
df[df['Country/Territory']== 'India']  # takes a single value only
```

#### By using 'isin()' function, which takes takes a list of values 

```python
df[df['Country/Territory'].isin(['Pakistan', 'India'])] #isin always takes a list [] inside for values
```

#### Using contain() function which is kind of similar to isin() but its more similar to 'like' in the SQL

```python
df[df['Country/Territory'].str.contains('Arab')]
```

## Filtering based on the Index

### Two different ways, using filter() and loc[] & iloc[]

#### Set index first, filtering based on index takes only works when dataframe is set on the mentioned index 

```python
df2 = df.set_index('Country/Territory')
```

```python
df2.filter(items = ['Pakistan'], axis = 0) # using filter with 'item' inside always takes a list [] inside and axis
```

```python
df2.filter(items = ['Pakistan', 'India'], axis = 0)
```

### Can also use 'like' inside the filter function

```python
df2.filter(like = 'Arab', axis = 0)    # like don't take list[]
```

#### loc[] always takes the current set_index value whether it is a string value or integer, use string with apostrophe. While                     iloc[] takes only defaut integer index, independly, of the string index.

```python
df2.loc['Pakistan'] 
```

```python
df2.iloc[0]
```

## Ordering 

### Ordering using sort by:   sort_values (by=   )

```python
df[df['Rank'] < 10 ].sort_values(by = 'Rank', ascending = False)
```

```python
df[df['Rank'] < 10 ].sort_values(by = ['Rank', 'Country/Territory'], ascending = [True, False] )
```
